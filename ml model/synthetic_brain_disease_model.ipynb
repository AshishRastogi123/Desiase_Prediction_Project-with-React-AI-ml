### #we will fill all nan values escape  Diagnosis           

```python
df['Gender']=df['Gender'].fillna('Unknown')
df['Education_Level']=df['Education_Level'].fillna('Not Given')
df['Headaches']=df['Headaches'].fillna('No')
df['Mood_Changes']=df['Mood_Changes'].fillna('No')
df['Mood_Changes']=df['Mood_Changes'].fillna('No')
df['Previous_Stroke']=df['Previous_Stroke'].fillna('No')
df['MRI_Abnormality']=df['MRI_Abnormality'].fillna('No')
```

```python
df['Family_History']=df['Family_History'].fillna('No')
```

```python
df = df.dropna(subset=['Diagnosis'])
```

```python
df.isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include="object").columns
categorical_cols
```

```python
numeric_cols = df.select_dtypes(include="number").columns
numeric_cols
```

## #Perorm Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), cmap="coolwarm", annot=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## #Prepare data for training

```python
x = df.drop("Previous_Stroke", axis=1)
y = df["Previous_Stroke"]
```

## #Train/test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## #Train Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

## #Evaluate the model

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print("Model Accuracy:", accuracy)
print("Classification Report:\n", report)
```
